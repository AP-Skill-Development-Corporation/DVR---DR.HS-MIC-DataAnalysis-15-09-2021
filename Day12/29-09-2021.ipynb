### Joint Plot
- Combination of two plots. By default it have scatterplot and histogram
    - sns.joinplot()

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'petal_length', y = 'sepal_length', data = iris, color = 'g')
```

```python
sns.jointplot(x = 'petal_length', y = 'sepal_length', data = iris, kind = 'hex', color = 'orange')
```

- KDE (Kernel Density Estimation)
    - Used to estimate probability density curve

```python
sns.jointplot(x = 'petal_length', y = 'sepal_length', data = iris, kind = 'kde', color = 'm')
```

### Pair plot
- It bulds a histogram and scatterplot for pairwise relationship
    - sns.paiplot()

```python
help(sns.pairplot)
```

```python
iris.head()
```

```python
sns.pairplot(iris)
## Diagnally(histogram) --> every column is related with itself
```

```python
## hue --> determines which column in dataset should be used for color encoding
sns.pairplot(iris, hue = "species")
```

```python
## for easily understanding
sns.pairplot(iris, hue = 'species', vars = ['sepal_length','petal_length'], palette = 'husl')
```

```python
sns.pairplot(iris, hue = 'species', markers = ['+','D','*'], palette = 'gnuplot')
```

```python
sns.pairplot(iris,hue='species', markers=['+','D','*'], palette='gist_gray_r')
```

```python
sns.pairplot(iris, hue = 'species', markers = ['+','D','*'], palette = 'gist_gray_r')
```

### Heatmap
- It displays matrix-like data. 
- Used to check, how strongly each column depended on other column
    - sns.heatmap()

```python
help(sns.heatmap)
```

```python
t = sns.load_dataset('tips')
t.head()
```

```python
sns.heatmap(t)
## Error: because of object datatype 
```

```python
t.iloc[:,[0,1,-1]].head()
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]])
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]].head())
```

```python
t.corr()
```

```python
sns.heatmap(t.corr(), annot = True)
```

```python
sns.heatmap(t.corr(), annot = True, cmap = 'Blue')
```

```python
sns.heatmap(t.corr(), annot = True, linewidths = 5,linecolor = 'pink',cmap = 'Pastel2')
```

### Lmplot()
- Combination of scatter and line plot

```python
t.head()
```

```python
help(sns.lmplot)
```

```python
sns.lmplot(x="total_bill", y="tip", data=t, palette = 'gist_gray_r')
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=t)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=t,palette = 'Set1')
## Palette will works after giving "hue" parameter
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=t, palette = 'Set1')
```

```python
sns.lmplot(x="total_bill", y="tip", hue = 'smoker', col="sex", data=t, palette = 'Set1')
```
