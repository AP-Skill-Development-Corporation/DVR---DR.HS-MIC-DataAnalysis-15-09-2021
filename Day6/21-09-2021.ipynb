## Data Cleaning Using Pandas
### Working on Duplicate values and Handling missing values

#### Identify and Removing Duplicate values in our data

```python
dict1 = {'Trainers':['Poojitha','Mounika','Lavanya','Prabhu','Poojitha'], 
         'Students':['Ramya','Durga','Likitha','Davoodh','Ramya']}
data = pd.DataFrame(dict1)
data
```

```python
data.duplicated()
```

```python
## identifying duplicate rows.
data[data.duplicated()]
```

```python
## Remove duplicates
data.drop_duplicates()
```

```python
data
```

- inplace = True -->  Changes will done in original dataset
- inplace = False --> No changes will done in original dataset.

```python
data.drop_duplicates(inplace = True)
```
