#### Identifying missing values and manipulate the data
- There are mainly 4 methods to identify & manipulating missing values
    - isnull() : Presence of missing values
    - notnull() : opposite to isnull()
    - dropna() : It delets the null values
    - fillna() : Replace the null values

```python
import pandas as pd
```

### Replace missing values
- If data is numerical, then replace null value either with '0' or with mean() 
- If data is categorical, then replace null value either with 'most occured value' or 'any object'
    - fillna()

```python
data2 = df.copy()
data2.head()
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data2['Product_Base_Margin'].mean()
```

```python
data2['Product_Base_Margin'] = data2['Product_Base_Margin'].fillna(data2['Product_Base_Margin'].mean())
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

### Encoding Techniques
- Encoding techniques is used to convert categorical data into numerical data
    - Label Encoder

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le
```

```python
titanic['Sex'] = le.fit_transform(titanic['Sex'])
```

```python
titanic.head()
```

```python
titanic['Cabin'] = le.fit_transform(titanic['Cabin'])
## Error: because of NaN values in Cabin column.
```
