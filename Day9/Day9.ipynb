```python
# Line plot
x = [11,12,13,14]
y = [1,2,3,4]
plt.plot(x,y)
plt.show()
```

```python
x = [10,11,12,13]
y = [1,2,3,4]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line plot")
#plt.legend(loc = 'lower right')
plt.plot(x,y)
plt.show()
plt.legend(loc = 'best')
```

```python
import numpy as np
x1 = np.arange(11,16)
y1 = np.arange(1,6)

x2 = np.linspace(10,25,5)
y2 = np.linspace(1,5,5)

plt.plot(x1,y1,color = "red",label = "Line1")
plt.plot(x2,y2,color = "green",label = "line2")

plt.legend()
```

```python
# bar chart
year = ["2019 Aug","2019 Dec","2020 Apr","2020 July","2020 Oct"]
goldprice = [1000,1500,3000,4000,5200]
plt.bar(year,goldprice,color = "c")
```

```python
year = ["2019 Aug","2019 Dec","2020 Apr","2020 July","2020 Oct"]
goldprice = [1000,1500,3000,4000,5200]
silver = [300,500,400,800,1000]
plt.bar(year,goldprice,color = "c")
plt.bar(year,silver,color = "m")
```

```python
x = [1,2,3,4,5]
y = [90,87,72,61,55]
plt.barh(x,y,color = ["r","g","y","k","m"],edgecolor = "blue")
```

```python
x = np.random.randint(100,200,50)
y = np.random.randint(1000,1500,50)
plt.scatter(x,y,marker = "P") # D,+
```

```python
# box plot
plt.boxplot(x)
```

```python
x = np.random.randint(100,200,50)
y = np.random.randint(100,500,100)

plt.boxplot([x,y],labels = ["x-axis","y-axis"],patch_artist = True,
            boxprops = dict(facecolor = "pink",color = 'r'),vert = False)
plt.show()
```
